```python
# – Impute categorical missing values with the mode.
dataset["Engine Fuel Type"].fillna(dataset["Engine Fuel Type"].mode([0]),inplace=False)
dataset["Market Category"].fillna(dataset["Market Category"].mode([0]),inplace=False)
dataset
```

# Use the IQR (Interquartile Range) method to quantitatively detect andoptionallyremoveoutliers.

```python
#Use the IQR (Interquartile Range) method to quantitatively detect and optionally remove outliers.
import numpy as np
import pandas as pd

a2=np.array(dataset['Engine HP'])
b1=np.array(dataset['highway MPG'])
c1=np.array(dataset['city mpg'])
d1=np.array(dataset['MSRP'])

sort_a2=np.sort(a2)
sort_b1=np.sort(b1)
sort_c1=np.sort(c1)
sort_d1=np.sort(d1)

Q1=np.percentile(sort_a2,25,interpolation='midpoint')
Q3=np.percentile(sort_a2,75,interpolation='midpoint')

IQR=Q3-Q1

low_bound=Q1-1.5*IQR
upp_bound=Q3+1.5*IQR
# print(low_bound,upp_bound)
outlier=[]
for x in sort_a2:
    if(x>upp_bound) or (x<low_bound):
     outlier.append(x)

print('Outlier of a1 is:',outlier)

Q1=np.percentile(sort_b1,25,interpolation='midpoint')
Q3=np.percentile(sort_b1,75,interpolation='midpoint')

IQR=Q3-Q1

low_bound=Q1-1.5*IQR
upp_bound=Q3+1.5*IQR
# print(low_bound,upp_bound)
outlier=[]
for x in sort_b1:
    if(x>upp_bound) or (x<low_bound):
     outlier.append(x)

print('Outlier of b1 is:',outlier)

Q1=np.percentile(sort_c1,25,interpolation='midpoint')
Q3=np.percentile(sort_c1,75,interpolation='midpoint')

IQR=Q3-Q1

low_bound=Q1-1.5*IQR
upp_bound=Q3+1.5*IQR
# print(low_bound,upp_bound)
outlier=[]
for x in sort_c1:
    if(x>upp_bound) or (x<low_bound):
     outlier.append(x)

print('Outlier of c1 is:',outlier)

Q1=np.percentile(sort_d1,25,interpolation='midpoint')
Q3=np.percentile(sort_d1,75,interpolation='midpoint')

IQR=Q3-Q1

low_bound=Q1-1.5*IQR
upp_bound=Q3+1.5*IQR
# print(low_bound,upp_bound)
outlier=[]
for x in sort_d1:
    if(x>upp_bound) or (x<low_bound):
     outlier.append(x)

print('Outlier of d1 is:',outlier)
ss=[sort_a2,sort_b1,sort_c1,sort_d1]
plt.boxplot(ss,tick_labels=["Engine HP", "highway MPG","city mpg",'MSRP'])
plt.savefig("myfig2.png")
plt.show()
```

# Summarize the key patterns observed in the data.

The given dataset contains information about various aspects, which include, specifically, car models, detailing their specifications and performance metrics.The dataset has 11914 rows and 16 columns.

The very first classification is about brand distribution:
Models from BMW, Audi, Fiat, Mercedes-Benz, Chrysler, and Nissan are the primary emphasis of the data, which is a compilation of several car specifications.

The second main key pattern observed is the model years.On going through the dataset, we can observe that these models of cars range from the early 1990s up to 2017.
The other main focus was brought upon the ‘Engine Specifications.’. This includes
1) Engine Fuel Type: Mostly premium unleaded and regular unleaded; some models also utilize diesel or flex-fuel systems.
2) Engine HP: Varies significantly, with values ranging from around 100 HP to over 300 HP depending on the model and year.
3) Engine Cylinders: Most vehicles have 4 to 6 cylinders.

The other key factor to catch our attention is "Vehicle Specifications."This includes:
1) Transmission Type: Transmission type advances from manual to automatic, generally with new models being welcomed with each coming year.
2) Driven_Wheels: It basically includes a proportionate mix of rear wheels, front wheels and all-wheel drivers
3) Number of Doors: Most vehicles have 2 to 4 doors.
4) Vehicle Types: Various styles, including sedans, coupes, convertibles, hatchbacks, and wagons, categorized into market categories like luxury, performance, and factory tuner.
5) Prices range significantly: from as low as 2,000 for older models to over 51,000 for high-performance or luxury models.

The next key patterns are based on “Performance Metrics.”
1) MPG Ratings: Highway MPG typically ranges from 25 to over 45, while city MPG varies around 15 to 32.
2) Popularity: Measured by the number of sales or other popularity metrics, with values provided where available.
3) Market Categories: Cars are categorized into several groups, such as Luxury, Performance, High-Performance, and more niche classifications, reflecting their positioning in the market

So this was the quick summary about the given dataset
